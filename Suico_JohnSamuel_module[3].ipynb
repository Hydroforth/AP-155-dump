**Physics Application: Rigid body rotations with matrices**

Consider the cube in this figure:

The cube is rotated with the three angular velocities:

$ \begin{equation} 
{\bf \omega} = 
\begin{pmatrix}

\begin{pmatrix}

1. Compute the angular momentum vector $\bf L$ via the matrix multiplication ${\bf L} = {\bf I\omega}$ for each case. (5 pts.)

2. Make three 3-D plots for ${\bf \omega}$ and ${\bf L}$ for each case. Label and describe your plots clearly. (10 pts.)

1. Here, I've initialized a 2-D array named `I` with elements given by the matrix $I$ above. I then initilialize each $\omega$ as `omega_1`, `omega_2`, and `omega_3`, respectively. I then multiply `I` with each $\omega$ using the `matmul` function and assign it to the variables `L_1`, `L_2`, and `L_3` respectively.

```python
I = array([[2/3, -1/4, -1/4],
         [-1/4, 2/3, -1/4],
         [-1/4, -1/4, 2/3]], float) 
omega_1 = array([1,0,0])
omega_2 = array([0,1,0])
omega_3 = array([1,1,1])

L_1 = matmul(I,omega_1)
L_2 = matmul(I,omega_2)
L_3 = matmul(I,omega_3)
```

2. Here, I've plotted each $\bf{\omega}$ and resulting $\bf{L}$ as a vector in 3 separate 3-D plots. I first speed up the process by creating lists to hold the matrices for $\bf{\omega}$ and $\bf{L}$ so I can loop through each and plot their values accordingly. I then use the `matplotlib` package to create separate 3-D plots, each with their own $\bf{\omega}$ and corresponding $\bf{L}$.

    From the plots, it can be seen that the smaller the angle between $\bf{\omega}$ and $\bf{L}$, the larger $\bf{L}$ is. This is consistent with the definition of angular momentum, since it is the cross product between the moment of inertia and the angular velocity.

```python
from mpl_toolkits import mplot3d
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

#initialize important variables
omega = [omega_1,omega_2,omega_3]
L = [L_1,L_2,L_3]
fig = []
ax = []

#make the plot
for i in range(0,3):
    
    #creating data points to plot
    omega_xyz = []
    L_xyz = []
    for j in range(0,3):
        omega_xyz.append(omega[i][j])
        L_xyz.append(L[i][j])
        
    #creating the figures    
    fig.append(plt.figure())
    ax.append(fig[i].add_subplot(111, projection = '3d'))
    
    #plotting the vectors
    ax[i].quiver(0,0,0,omega_xyz[0],omega_xyz[1],omega_xyz[2], color = 'red')
    ax[i].quiver(0,0,0,L_xyz[0],L_xyz[1],L_xyz[2], color = 'blue')
    
    #labelling the vectors 
    omega_label = r'$\omega$' 
    L_label = r'$L$'
    ax[i].text(omega_xyz[0],omega_xyz[1],omega_xyz[2], omega_label, color='black')
    ax[i].text(L_xyz[0],L_xyz[1],L_xyz[2], L_label, color='black')
    
    #setting axis limits
    ax[i].set_xlim(-1,1)
    ax[i].set_ylim(-1,1)
    ax[i].set_zlim(-1,1)
    plotTitle = "L_" + str(i+1) + " and $\omega$_" + str(i+1) + " vectors" 
    
    #labelling the plot and axes
    plt.title(plotTitle)
    ax[i].set_xlabel('x axis')
    ax[i].set_ylabel('y axis')
    ax[i].set_zlabel('z axis')
```

**Solve a nonlinear equation using the relaxation method**

Consider the equation $x=1- e^{-cx}$, where $c$ is a known parameter and $x$ is unknown. This equation arises in a variety of situations, including the physics of contact processes, mathematical models of epidemics, and the theory of random graphs.

1. Write a program to solve this equation for $x$ using the relaxation method for the case $c = 2$. Calculate your solution to an accuracy of at least $10^{-6}$. *(10 pts.)*

2. Modify your program to calculate the solution for values of $c$ from 0 to 3 in steps of 0.01. *(10 pts.)* 

3. Make a plot of $x$ as a function of $c$. Describe the different regimes that you see. *(5 pts.)*

1. Here, I've defined a function `relax` which takes 2 arguments `c` and `accuracy`. The `relax` function uses the relaxation method and calculates a solution to the equation $$ x = 1 - e^{-cx}$$ to an accuracy given by the argument passed on to `accuracy`. 

```python
from math import exp
from numpy import abs

def relax(c,accuracy):
    save = 0
    x = 1
    while(abs(x-save) > accuracy):
        save = x
        x = 1 - exp(-c*x)
    return x
```

2. Here, I've created two variables `c` and `x`. The first variable `c` is an array of values generated by the `arange` function from $0$ to $3$ in steps of $0.01$. I then use a `for` loop to pass each value in `c` as an argument to the `relax` function, each with an accuracy of $10^{-6}$. I then plot the values of `x` and `c`, specifically `x` as a function of `c`. 

    From the graph, the two regimes are $0 \leq c \leq 1$ and $c > 1$. For the first regime, the values in `x` are constantly equal to $0$. However, for $c > 1$, the values of `x` begin to take increasing non-zero values, with decreasing increments as the values in `x` get closer and closer to the solution.

```python
from numpy import arange
import matplotlib.pyplot as plt

c = arange(0,3.01,0.01)
x = []
for i in c:
    x.append(relax(i,10**-6))
    
fig = plt.figure()
ax = fig.add_subplot(111)

ax.set_xlabel("c")
ax.set_ylabel("x")

plt.plot(c,x)
plt.title("Plotting x as a function of c")
```

**Physics Application: Calculating the Lagrange point**

_Refer to Exercise 6.16 of the Newman text._ The $L_1$ Lagrange point is the point between the Earth and the Moon at which a satellite will orbit the Earth in perfect synchrony with the Moon, staying always in between the two. 

Assuming circular orbits, and assuming that the Earth is much more massive than either the Moon or the satellite, the distance $r$ from the center of the Earth to the $L_1$ point satisfies
$$ \begin{equation}
\frac{GM}{r^2} - \frac{Gm}{(R-r)^2} = \omega^2 r,
\end{equation} $$
where $R$ is the distance from the Earth to the Moon, $M$ and $m$ are the Earth and Moon masses, $G$ is Newton's gravitational constant, and $\omega$ is the angular velocity of both the Moon and the satellite.

Use (i) Newton's method and (ii) the secant method to solve for the distance r from the Earth to the $L_1$ point. Compute a solution accurate to at least four significant figures. (15 pts. each)

1. Here, I've defined the function `poly` which returns the value y for the equation $$ y = \omega^2r - \frac{GM}{r^2} + \frac{Gm}{(R-r)^2}$$ given $r$. I then defined a function `polyprime` as the derivative of `poly` with respect to $r$.

    Afterwards, I defined `newt` as a function with the parameters `x`, `accuracy`, `f`, and `fprime`. The `newt` function takes on an initial guess `x` and calculates the root of a function using Newton's method. That is, we calculate a better guess $x_{n+1}'$ using the equation $$x_{n+1} = x_{n} - \frac{f(x)}{f'(x)}$$ The `f` and `fprime` parameters are the function and derivative of the function with respect to $x$, respectively.
    
    Similarly, I defined a function `secant` with parameters `x_1`, `x_2`, `accuracy`, `f` and `fprime`. The `newt` function takes on two initial guesses `x_1` and `x_2` and calculates the root of a function using the secant method. That is, we calculate a better guess $x_{3}$ using the equation $$x_3 = x_2 - f(x_2)\frac{x_2 - x_1}{f(x_2) - f(x_1)}$$ The `f` and `fprime` parameters are the function and derivative of the function with respect to $x$, respectively.
    
    I then test out both functions using an initial guess of $-2*10^{10}$ for the `newt` function and $-1 \times 10^{10}$ and $-1 \times 10^{9}$ for the `secant` method, both up to an accuracy of $1^{-6}$ to calculate a Lagrange point between the Earth and the Moon.

```python
def poly(r):
    G = 6.674 * 10**-11
    M = 5.974 * 10**24
    m = 7.348 * 19**22
    R = 3.844 * 10*8
    omega = 2.662 * 10**-6
    return omega**2*r - G*M*r**(-2) + G*m*(R-r)**(-2)
    
def polyprime(r):
    G = 6.674 * 10**-11
    M = 5.974 * 10**24
    m = 7.348 * 19**22
    R = 3.844 * 10*8
    omega = 2.662 * 10**-6
    return omega**2 + 2*G*M*r**(-3) + 2*G*m*(R-r)**(-3)
    
def newt(x, accuracy,f,fprime):
    prev = 1
    while(abs(f(x)-f(prev)) > accuracy):
        prev = x
        x = x - (f(x)/fprime(x))
    return x

def secant(x_1,x_2, accuracy, f, fprime):
    while(abs(f(x_2) - f(x_1))>accuracy):
        save = x_2
        x_2 = x_2 - f(x_2)*( (x_2-x_1)/(f(x_2) - f(x_1)) )
        x_1 = save
    return x_2

print("One solution given by Newton's method: ", newt(30000000,1**-8,poly,polyprime))
print("One solution given by the secant method: ", secant(3*10**10,30000000,1**-6,poly,polyprime))
```
