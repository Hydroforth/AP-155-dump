**Fibonacci sequence**

1. Write a function that returns the nth Fibonacci number. *(10 pts.)*

2. Print the nth Fibonacci number, for $n = 1, 10, 10^3, 10^6, 10^9$. *(5 pts.)*

1. Here I've written a function called `fibo` which takes an argument $n$ and returns the $n$th number from the Fibonacci Sequence. <br><br> 

    The Fibonacci sequence is defined such that the $n$th number is the sum of the preceding two numbers $n-1$ and $n-2$, starting from $0$ and $1$. I've defined two separate variables to function as the starting points for the sequence: `a_0` with the value $0$ and `a_1` with the value $1$. <br><br> 

    To continue the sequence, I update the value of `a_1` while <b>simultaneously</b> updating the value of `a_0` to the <b>previous</b> value of `a_1` (I had previously solved the problem by creating a dummy variable to temporarily 'hold on' to the <b>previous</b> value of `a_1`; this approach is simplified by Python). After looping $n$ times, the value of `a_0`, which corresponds to the value of the $n$th number in the Fibonacci sequence, is returned.

````python
def fibo(n):
    a_0,a_1 = 0,1
    for i in range(n):
        a_1,a_0 = (a_1 + a_0),a_1
    return a_0
````

2. Here I've written two functions called `ordinalIndicator` which returns the appropriate ordinal indicator for a number, and `output` which prints the $n$th number of the Fibonacci Sequence. <br><br>
    
    The function `ordinalIndicator` takes an (assumingly) int or float type argument $n$ and converts the said number into a string. I make use of Python's indexing of strings to pinpoint the number in the argument's <b>ones place</b> and subsequently return the appropriate ordinal indicator. I only wrote this argument to improve the readability of function `output`. Note that this function only works for integers.<br><br>
    
    The function `output` takes an argument $n$ and prints the string "This is the $n$<i>(ordinal indicator)<sub>[1]</sub></i> number in the Fibonacci Sequence: <i>($n$th number in the Fibonacci Sequence)<sub>[2]</sub>"</i>
    
    <sub>[1] The ordinal indicator is determined by passing the argument $n$ to the function `ordinalIndicator`. </sub> <br>
    <sub>[2] The $n$th number in the Fibonacci Sequence is determined by passing the argument $n$ to the function `fibo`.</sub>
    
    I subsequently call the `output` function 4 times with the following arguments, respectively: $1$, $10$, $1000$, $1000000$.

````python
def ordinalIndicator(n):
    convert = str(n)
    onesPlace = int(convert[len(convert)-1])
    
    if onesPlace == 0 or onesPlace>3:
        return "th"
    elif onesPlace == 1:
        return "st"
    elif onesPlace == 2:
        return "nd"
    elif onesPlace == 3:
        return "rd"
    
def output(n): 
    print("This is the ",n,ordinalIndicator(n)," number in the Fibonacci Sequence: ",fibo(n),sep="")

output(1)
output(10)
output(10**3)
output(10**6)
````

````python
from mpl_toolkits import mplot3d
%matplotlib inline
from numpy import cos,sin,pi,linspace, random
import matplotlib.pyplot as plt

class circularHelix:
    def __init__(self):
        self.R,self.a = 10,3
        self.t = linspace(0,4*pi,100)
        self.x = self.R*cos(self.t)
        self.y = self.R*sin(self.t)
        self.z = self.a*self.t
    
    def drawCurve(self,doMore=False):        
        fig = plt.figure()
        ax = plt.axes(projection = '3d')
        ax.plot3D(self.x,self.y,self.z, 'gray')
        if doMore == True:
            return ax
    
    def plotRandomonCurve(self,doMore=True):
        ax = self.drawCurve(doMore)
        
        #generates erratic data points
        t_rand = 4*pi * random.random(100)
        x_rand = self.R*cos(t_rand) + 1.5*random.random(100)
        y_rand = self.R*sin(t_rand) + 1.5*random.random(100)
        z_rand = self.a*t_rand
        
        ax.scatter3D(x_rand,y_rand,z_rand,cmap='Greens')
    
circularHelix().drawCurve()
````

The float type variables `uf_limit` and `of_limit` are used to save each iteration of `uf_counter` and `of_counter`, respectively, before updating each iterator variable's values by a factor of 2<sub>[2]</sub>. As such, when `uf_counter` and  `of_counter` hit their respective limits <b>0.0</b> and <b>inf</b>, the values for the underflow and overflow limits are retained. The class also initializes two variables `uf_loop` and `of_loop` which are used to count the number of loops until underflow and overflow occurs, respectively. <br><br> <sub>[1] I wrote a class because I didn't want to make two separate functions to calculate the underflow and overflow limits.</sub><br>

**Calculating the sum of a simple series**

In spite of the power of your trusty computer, calculating the sum of even a simple series may require some thought and care. Consider the two series

$$\begin{equation}
S^{\rm (up)} = \sum_{n=1}^{N} \frac{1}{n},\ S^{\rm (down)} = \sum_{n=N}^{1} \frac{1}{n}.
\end{equation}
$$

Both series are finite as long as $N$ is finite, and when summed analytically both give the same answer. Nonetheless, because of round-off error, the numerical value of $S^{\rm (up)}$ will not be precisely that of $S^{\rm (down)}$.

1. Calculate $S^{\rm (up)}$ and $S^{\rm (down)}$ as functions of $N$. *(10 pts.)*

2. Make a log-log plot of $(S^{\rm (up)}-S^{\rm (down)})/(|S^{\rm (up)}|+|S^{\rm (down)}|)$ vs. $N$. *(10 pts.)*

3. Observe the linear regime on your graph and explain why the downward sum is generally more precise. *(5 pts.)*

1. Here, I've written a class called `sumSeries` with the instance variable `N` which is initialized as itself. I've also written multiple functons under the class which I will explain briefly.<br><br>

    The function `genericUp` takes an argument $n$ and then returns the sum of the series given above. The calculation is facilitated through a `for` loop with the iterating variable `i` initialized at $1$ with a $+1$ increment. Each iteration adds the value  $\frac{1}{i}$ to the variable `sumUp`, until `i` equals $n$ (thereby calculating $S^{\rm (up)}$). The function `genericDown` also returns the sum of the series given above, however, the `for` loop's iterating variable `i` is initialized with the value `n` and instead decrements by `-1`. Each iteration adds the value  $\frac{1}{i}$ to the variable `sumDown`, until `i` equals $1$ (thereby calculating $S^{\rm (down)}$). <br><br>
    
    The functions `specificUp` and `specificDown` also calculate the sum of the series $\frac{1}{n}$, except that they are specifically used to calculate the sums using the instance variable `N`. The said functions actually only pass the argument `N` to the `genericUp` and `genericDown` functions.<sub>[1]</sub>
    
    <sub>[1] The `specificUp` and `specificDown` functions are a bit redundant and are only used to answer Problem no. 1</sub> <br><br>
    
    I first initialize an instance of the `sumSeries` class called `testme`. I then print out the string "S(up): $S^{\rm (up)}$"<sub>[2]</sub> on the first line, and then the string "S(down): $S^{\rm (down)}$<sub>[3]</sub>" on the next line.
    
    <sub>[2] $S^{\rm (up)}$ is calculated by calling the class function `specificUp` using the class instance `testme` for $N = 1000$</sub> <br>
    <sub>[3] $S^{\rm (down)}$ is calculated by calling the class function `specificDown` using the class instance `testme` for $N = 1000$</sub>
    

````python
%matplotlib inline
import matplotlib.pyplot as plt
from numpy import abs,linspace

class sumSeries:
    def __init__(self,N):
        self.N = N
    
    def genericUp(self,n):
        sumUp = 0
        for i in range(1,n+1):
             sumUp += 1/i
        return sumUp
    
    def genericDown(self,n):
        sumDown = 0
        for i in range(n,0,-1):
            sumDown += 1/i
        return sumDown
    
    def specificUp(self):
        return self.genericUp(self.N)
    
    def specificDown(self):
        return self.genericDown(self.N)
    
    def makePlot(self):
        y = list()
        
        #calculating x:
        x = linspace(1,self.N,self.N)
        
        #calculating y:
        for i in range(1,self.N+1):
            y.append( abs( (self.genericUp(i) - self.genericDown(i)) / ( abs((self.genericUp(i)) + abs(self.genericDown(i))) ) ))
        
        plt.xscale("log")
        plt.yscale("log")             
        plt.scatter(x,y)
 
testme = sumSeries(1000)
print("S(up): ",testme.specificUp())
print("S(down): ",testme.specificDown())
#testme.makePlot()
    
````

3. Addition and subtraction using float type variables tends to produce a rounding error. The graph shows that as $N$ increases, the difference between $S^{\rm (up)}$ and $S^{\rm (down)}$ increases, thereby implying that the rounding error accumulates as $N$ increases. The downward sum generally means that the computer will be calculating the sum by adding smaller numbers initially, until $N = 1$. This means that the relative difference between each consecutive term $\frac{1}{i}$ will be smaller than that of the upward sum (i.e. which starts with larger relative differences). The smaller relative differences mean that there is less room for rounding error in the downwards approach (i.e. because of the similarity in their least significant digits). These errors accumulate across each operation, but the errors accumulate much faster in the upwards approach, hence the rising peaks in the graph as $N$ increases.
